3次関数+正規分布ノイズに従うデータを生成し、そのデータに対し1次,2次,3次の近似式を作成します。

```python
# ランダムシードの指定
np.random.seed(1483679)

# オリジナルデータの生成
x = np.linspace(0.5, 3.5, 20)
y = (x**3 + x**2 + x + np.random.randn(20)*10) + 2
y = 4*y/(y.max())

#近似式の係数
coe1 = np.polyfit(x, y, 1)
coe2 = np.polyfit(x, y, 2)
coe3 = np.polyfit(x, y, 3)

#近似式の計算
y1 = np.poly1d(coe1)(x) #1次
y2 = np.poly1d(coe2)(x) #2次
y3 = np.poly1d(coe3)(x) #3次
```

⓪図の操作を行うためのインスタンスを作成します。

```python
# ------------------------------------------------------------------------------------------------------------------------------------------------------

# ⓪インスタンスの生成
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)

# ------------------------------------------------------------------------------------------------------------------------------------------------------
```

①生成したデータとその近似関数をプロットします。

```python
# ⓪インスタンスの生成
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)


# ------------------------------------------------------------------------------------------------------------------------------------------------------

# ①データのプロット
ax.plot(x, y1, c='r', lw=2, label="first approximation", zorder=10)
ax.plot(x, y2, c='g', lw=2, linestyle='-', label="second approximation", zorder=10)
ax.plot(x, y3, c='b', lw=2, linestyle='--', label="third approximation")
ax.plot(x, y, linewidth=0, marker='o', markerfacecolor='w', markeredgecolor='k', label="元のデータ")

# ------------------------------------------------------------------------------------------------------------------------------------------------------
```

②図のタイトルをつけます。

```python
# ⓪インスタンスの生成
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)

# ①データのプロット
ax.plot(x, y1, c='r', lw=2, label="first approximation", zorder=10)
ax.plot(x, y2, c='g', lw=2, linestyle='-', label="second approximation", zorder=10)
ax.plot(x, y3, c='b', lw=2, linestyle='--', label="third approximation")
ax.plot(x, y, linewidth=0, marker='o', markerfacecolor='w', markeredgecolor='k', label="元のデータ")


# ------------------------------------------------------------------------------------------------------------------------------------------------------

# ②図のタイトル
ax.set_title("Title of figure", fontsize=20)

# ------------------------------------------------------------------------------------------------------------------------------------------------------
```

## オプション

#### ⓪インスタンス

    figsize : 図の大きさの設定。figsize=(横, 縦)により、インチ単位で指定できます。
    
        Ex)　fig = plt.figure(figsize=(8, 8))

    aspect : 図のアスペクト比の設定。aspect=縦/横を指定できます。
        Ex) ax = fig.add_subplot(1, 1, 1, aspect=0.5)
             この場合は、縦：横=1:2となる。

    ※figsizeを指定している場合、aspectは不要。（figsizeの比がアスペクト比となります。）

#### ①プロット

    color : プロットの色を設定します。以下にて色を指定できます。
    
        ◾️単色
            'b'     青 (Blue)
            'g'     緑 (Green)
            'r'      赤 (Red)
            'c'     シアン (Cyan)
            'm'    マゼンタ (Magenta)
            'y'     黄 (Yellow)
            'k'     黒 (Black)
            'w'    白 (White)

        ◾️グレースケール
            0.0（黒）〜1.0（白）

    label :　プロットの凡例の表示名を設定します。

    linewidth : ラインプロットの幅を設定します。

    linestyle : ラインプロットの種類を設定します。以下にて指定できます。
    
        '-'   ： 実線
        'ｰｰ' ： 破線
        '-.'  ： 一点鎖線
        ':'   ： 点線

    zorder : プロットの深度を設定します。値が大きいほど前面にプロットされます。
        (平面のxy軸に対するz軸と考えてください。画面から手前がz軸正方向、奥がz軸負方向です。)

    marker : マーカープロットの種類を設定します。以下にて指定できます。
        
        ‘.’ , ‘o’ , ‘x’, ‘+’ , ‘^’ , ‘*’

    markerfacecolor : マーカープロットの塗りつぶし色を設定します。先述のcolorオプションと同様の指定ができます。

    markeredgecolor : マーカープロットの縁取り色を設定します。先述のcolorオプションと同様の指定ができます。

        Ex)ax.plot(x, y,　marker='o', markerfacecolor=‘w', markeredgecolor='k')
            白抜き黒縁取り円のマーカープロットとなります。

#### ⑧⑨目盛

    which : 設定を行う目の種類を指定します。
        'major' : 大目盛
        'minor' : 小目盛

    labelsize : 目盛の文字サイズを設定します。

    labelcolor : 目盛の文字色を設定します。先述のcolorオプションと同様の指定ができます。

    width : 目盛線の幅を設定します。

    length : 目盛線の長さを設定します。

#### ⑩注釈

    labelsize : 注釈テキストの文字サイズを設定します。

    ha : 注釈の横方向の挿入位置を設定します。
        　　　第1,2引数で指定した挿入座標(x,y)にテキストの左右端どちらを揃えるかを指定できます。
            'right' : 右揃え
            'left' 　　: 左揃え

    va : 注釈の縦方向の挿入位置を設定します。
            'top'    　　　: 上揃え
            'bottom' :　下揃え
            '

#### ①サブプロット

```python
# インスタンスの生成
fig = plt.figure(figsize=(6, 6))

# ------------------------------------------------------------------------------------------------------------------------------------------------------

# 2図を横に並べて表示
# add_subplot(x,y,z)   x : 行数、y : 列数、z : 表示位置
# sharex : 指定した図のx軸を共有　　　sharey : 指定した図のy軸を共有
ax1 = fig.add_subplot(1, 2, 1, aspect='2')
ax2 = fig.add_subplot(1, 2, 2, aspect='2', sharey=ax1)

# ------------------------------------------------------------------------------------------------------------------------------------------------------

# データのプロット
# データは単数プロットと同様のものを使用
ax1.plot(x, y1, c='r', lw=2, label="first approximation", zorder=10)
ax1.plot(x, y2, c='g', lw=2, linestyle='-', label="second approximation", zorder=10)
ax1.plot(x, y3, c='b', lw=2, linestyle='--', label="third approximation")
ax2.plot(x, y, linewidth=0, marker='o', markerfacecolor='w', markeredgecolor='k', label="original data")


## 以下、各図に対して設定を行います。 ##


# 図のタイトル
ax1.set_title("Title of figure1", fontsize=15, verticalalignment='bottom')
ax2.set_title("Title of figure2", fontsize=15, verticalalignment='bottom')

#凡例
ax1.legend()
ax2.legend()

# X軸ラベル
ax1.set_xlabel("X axis label")
ax2.set_xlabel("X axis label")

#　Y軸ラベル
ax1.set_ylabel("Y axis label")
ax2.set_ylabel("Y axis label")

#　グリッド
ax1.grid(linestyle="--", linewidth=0.5, color='.25', zorder=-10)
ax2.grid(linestyle="--", linewidth=0.5, color='.25', zorder=-10)

#　X軸目盛の設定
ax1.xaxis.set_major_locator(MultipleLocator(1.000))
ax2.xaxis.set_major_locator(MultipleLocator(1.000))
ax1.xaxis.set_minor_locator(AutoMinorLocator(4))
ax2.xaxis.set_minor_locator(AutoMinorLocator(4))

#　Y軸目盛の設定
ax1.yaxis.set_major_locator(MultipleLocator(1.000))
ax2.yaxis.set_major_locator(MultipleLocator(1.000))
ax1.yaxis.set_minor_locator(AutoMinorLocator(4))
ax2.yaxis.set_minor_locator(AutoMinorLocator(4))

#　X軸の値範囲
ax1.set_xlim(0, 4)
ax2.set_xlim(0, 4)

#　Y軸の値範囲
ax1.set_ylim(0, 4)
ax2.set_ylim(0, 4)
```

```python
# インスタンスの生成
fig = plt.figure(figsize=(6, 6))

# ------------------------------------------------------------------------------------------------------------------------------------------------------

# 1つの図を分割して、2図を縦に並べて表示
# add_axes(x,y,x-ratio,yratio)
# x,y : fig内のxy座標を指定し、図の原点を合わせます。
# x-ratio : fig内のX軸の比率、　y-ratio : fig内のY軸の比率
ax1 = fig.add_axes((0, 0.2, 1, 0.8))
ax2 = fig.add_axes((0, 0, 1, 0.2), sharex=ax1)

# ------------------------------------------------------------------------------------------------------------------------------------------------------

# データのプロット
# データは単数プロットと同様のものを使用
ax1.plot(x, y1, c='r', lw=2, label="first approximation", zorder=10)
ax1.plot(x, y2, c='g', lw=2, linestyle='-', label="second approximation", zorder=10)
ax1.plot(x, y3, c='b', lw=2, linestyle='--', label="third approximation")
ax2.plot(x, y, linewidth=0, marker='o', markerfacecolor='w', markeredgecolor='k', label="original data")


## 以下、各図に対して設定を行います。 ##


# 図のタイトル
ax1.set_title("Title of figure", fontsize=20, verticalalignment='bottom')

#凡例
ax1.legend()
ax2.legend()

# X軸ラベル
ax2.set_xlabel("X axis label")

#　Y軸ラベル
ax1.set_ylabel("Y axis label")

#　グリッド
ax1.grid(linestyle="--", linewidth=0.5, color='.25', zorder=-10)
ax2.grid(linestyle="--", linewidth=0.5, color='.25', zorder=-10)

#　X軸目盛の設定
ax2.xaxis.set_major_locator(MultipleLocator(1.000))
ax2.xaxis.set_minor_locator(AutoMinorLocator(4))

#　Y軸目盛の設定
ax1.yaxis.set_major_locator(MultipleLocator(1.000))
ax1.yaxis.set_minor_locator(AutoMinorLocator(4))

#　X軸の値範囲
ax1.set_xlim(0, 4)
ax2.set_xlim(0, 4)

#　Y軸の値範囲
ax1.set_ylim(0, 4)


# ------------------------------------------------------------------------------------------------------------------------------------------------------

# X軸 / Y軸目盛の非表示設定
# 上図のX軸目盛、下図のY軸目盛が被ってしまうため、非表示とします。
ax1.tick_params(labelbottom=False)
ax2.tick_params(labelleft=False)

# ------------------------------------------------------------------------------------------------------------------------------------------------------
```

#### ・オブジェクトの非表示

    プロットコマンドの末尾に；（セミコロン）をつけることで、オブジェクトを非表示にできます。
    今回は、先のヴァイオリンプロットを例に説明します。

```python
# データの生成
np.random.seed(1483680)
x1 = np.random.randn(300) + 1

np.random.seed(1483681)
x2 = np.random.randn(300) -1

x = (x1, x2)

# インスタンスの生成
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)

# ヴァイオリンプロット
# セミコロンをつけないとオブジェクト表示
ax.violinplot(x)
```
